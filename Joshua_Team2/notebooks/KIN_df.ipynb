```python
import demoji
import re

def clean(text):
  text = text.replace('@user', '')
  text = demoji.replace(text, '')
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  text = re.sub(r'https?://\S+', '', text)

  return text

print(clean(kin_df['tweet'].iloc[0]))
```

```python
kin_df['tweet'] = kin_df['tweet'].apply(clean)
```
